```python
from fastai import *
from fastai.vision import *
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import io
from PIL import Image
```

```python
model_path = Path('../../challenges/2/images')
model_path
```

## Importando Modelo

```python
learn = load_learner(model_path, 'resnet50.pkl')
```

```python
images_path = '../../neural-network-api/public/'
```

### Criando método para ler imagens em byte

```python
def read_byte_img(path):
    img = Image.open(path, mode='r')
    imgByteArr = io.BytesIO()
    img.save(imgByteArr, format='PNG')
    
    return imgByteArr
```

## Cabo em condições normais

```python
img_to_show = mpimg.imread(images_path + '1.png')
```

```python
imgplot = plt.imshow(img_to_show)
```

```python
byte_img = read_byte_img("../../neural-network-api/public/1.png")
```

```python
img = open_image(byte_img)
```

### Testando o modelo

```python
learn.predict(img)[0]
```
